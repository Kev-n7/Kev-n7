```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
%matplotlib inline
from gensim.parsing.preprocessing import remove_stopwords
import string
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10, 7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
```

```python
data = pd.read_csv('C:/Users/kev/Downloads/Fruit_Prices_2020.csv')
data.head()
```

```python
data.info()
```

```python
# Checking if the columns have null values in the dataset
data.isnull().sum()
```

```python
# Drop unwanted names in column
data.drop(['Unnamed: 8','Unnamed: 9', 'Unnamed: 10'],axis=1,inplace=True)
data.head()
```

```python
# Rename names in column
data.rename({'Fruit':'fruit', 'Form':'form', 'RetailPriceUnit':'retailpriceunit', 'RetailPrice':'retailprice', 'Yield':'yield', 'CupEquivalentSize':'cupequivalentsize', 'CupEquivalentUnit':'cupequivalentunit', 'CupEquivalentPrice':'cupequivalentprice'},axis=1,inplace=True)
data.head()
```

```python
# Create Profit variable multiplying '*' it by yield 
data['profit'] = data['retailprice'] * data['yield']
data.head(20)
```

```python
# Check for fruits value count
fruits = data.fruit.value_counts().sort_values(ascending=False).head(10)
fruits
```

```python
# Products sold
plt.figure(figsize=(12,6))
plt.title('Most Product Sold')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation = 90, fontsize = 20)
plt.hist(data.form, color='green', edgecolor = 'black');
```

```python
#plotting a histogram for fruits sold
data.fruit.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Fruits Sold')
plt.ylabel('')
plt.xlabel('');
```

```python
# value count for cupequivalentsize
data['cupequivalentsize'].value_counts()
```

```python
# Checking for unique fruits
data['fruit'].unique()
```

```python
# Apricots fruit yielded most profit in 2020
profit = data.groupby('fruit')['profit'].sum()
profit
```

```python
# Check fruits sum by 'retailpriceunit' and 'cupequivalentunit' 
profit = data.groupby('fruit')['retailpriceunit','cupequivalentunit'].sum()
profit
```

```python
# Fruits and drinks with retail price
plt.figure(figsize=(12,6))
plt.title('Fruits And Drink With Most Profit', size = 20)
plt.xticks(rotation = 30, ha='right')
sns.barplot(data.fruit, data.profit, palette = 'Set2');
```
