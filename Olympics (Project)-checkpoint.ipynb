```python
womenOlympics =  athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')]
```

```python
# Plotting a count plot for number of feamle participants in the summer olympics

sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlympics, palette='Spectral')
plt.title('Women Participation')
```

```python
# Plotting a line graph to show the trend of female participation since it's inception

part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes Over Time')
```

* There was a slight decrease in women participation in the olympics in the 1950's and 1980's but since then there has been a massive increase in female participation till date.

```python
# Gold medals won by athletes

goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals.head()
```

```python
# Taking only the values that are different from NaN.

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# Athletes that have secured a gold medal that are above the age of 60

goldMedals['ID'][goldMedals['Age'] > 60].count()
```

* Only 6 Athletes above the age of 60 have won a Gold medal in the olympics

```python
# Sport events that the 6 athletes won a Gold medal

sporting_event = goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

* Art competition, Roque, Archery and Shooting are the four events which the 6 athletes who are above the age of 60 won in the olympics with Archery as the most won of the four sports.

```python
# Plotting a count plot for the sporting_event 

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals For Athletes Over 60 Years')
```

```python
# Gold medals from each country

goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
```

* The United States has won more medals than any other country who has participated in the olympics since it's inception followed by Russia, Germany, UK, Italy and France.

```python
# Gold medals from each country

goldMedals.Region.value_counts().reset_index(name='Medal').tail(6)
```

* Countries like the Ivory coast, Bahrain, Grenada, Jordan, Vietnam and Mozambique have only managed to pick up just a single meadal in the olympics since it's inception.

```python
# Plotting a graph for Gold medals from each country

totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x='index', y='Medal', data=totalGoldMedals, height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 Countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per Country')
```

```python
# First Olympics Event 1896

min_year = athletes_df.Year.min()
print(min_year)

team_names = athletes_df[(athletes_df.Year == min_year) & (athletes_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

* In it's inception in '1896' Germany had the most Gold medals with '24'Gold medals with the United States coming second with '11' Gold medals and Greece coming very close with '10' Gold medals.

```python
# Rio Olympics Events 2016

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

* In It's latest Olympics event in Rio the United States had the most Gold medals of '137' and Great Britain coming in second place with '64', Russia with '50', Germany coming fourth with '47', China with '44', Brazil with '34', Australia with '23', Argentina with '21', France with '20' and Japan with '17' Gold medals.

```python
# Plotting a Barplot for the Rio Olympics

sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Countrywise Medals for the year 2016');
```
